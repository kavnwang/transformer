```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
loss_curve = pd.read_csv("models/test/stats.csv")
```

```python
plt.plot(loss_curve["loss"])
plt.xlabel("Batch")
plt.ylabel("Loss")
```

```python
loss_curve2 = pd.read_csv("models/test2/stats.csv")
```

```python
fig, ax = plt.subplots()
ax2 = plt.twinx(ax)
ax.plot(loss_curve2["loss"])
ax.set_xlabel("Batch")
ax.set_ylabel("Loss")
ax2.plot(loss_curve2["lr"])
ax2.set_ylabel("Learning Rate")
```

```python
fig, ax = plt.subplots()
ax2 = plt.twinx(ax)
ax.plot(loss_curve2["loss"], alpha=0.5)
ax.plot(loss_curve["loss"], alpha=0.5)
ax.set_xlabel("Batch")
ax.set_ylabel("Loss")
ax2.plot(loss_curve2["lr"])
ax2.set_ylabel("Learning Rate")
```

```python
loss_curve3 = pd.read_csv("models/test3/stats.csv")
```
